```python
# Define an empty list to store the tasks
tasks = []

# Define a function to add a task to the list
def add_task():
    task = input("Enter the task: ")
    tasks.append(task)
    print(f"Task '{task}' added to the list.")

# Define a function to remove a task from the list
def remove_task():
    task = input("Enter the task to remove: ")
    if task in tasks:
        tasks.remove(task)
        print(f"Task '{task}' removed from the list.")
    else:
        print(f"Task '{task}' not found in the list.")

# Define a function to display the list of tasks
def show_tasks():
    if len(tasks) == 0:
        print("The list is empty.")
    else:
        print("List of tasks:")
        for i, task in enumerate(tasks):
            print(f"{i+1}. {task}")

# Define a function to clear the list of tasks
def clear_tasks():
    tasks.clear()
    print("The list has been cleared.")

# Define a function to display the menu of options
def show_menu():
    print("Menu:")
    print("1. Add task")
    print("2. Remove task")
    print("3. Show tasks")
    print("4. Clear tasks")
    print("5. Exit")

# Define the main function to run the app
def main():
    while True:
        show_menu()
        choice = input("Enter your choice: ")
        if choice == "1":
            add_task()
        elif choice == "2":
            remove_task()
        elif choice == "3":
            show_tasks()
        elif choice == "4":
            clear_tasks()
        elif choice == "5":
            print("Goodbye!")
            break
        else:
            print("Invalid choice. Try again.")

if __name__ == "__main__":
    main()
```
